```python
df["Claim_Submission_Date "] = pd.to_datetime(df["Claim_Submission_Date"] )
```

```python
df["Claim_Submission_Date "].isnull().sum()
```

```python

cat_cols = ['Patient_Gender', 'Insurance_Type', 'Claim_Status', 'Visit_Type', 'Provider_Specialty', 'Patient_State']
for col in cat_cols:
    print(col, ":", df[col].unique())
```

```python
invalid_age = df[(df['Patient_Age'] < 0) | (df['Patient_Age'] > 120)]
invalid_amount = df[df['Claim_Amount'] <= 0]
approved_exceeds = df[df['Approved_Amount'] > df['Claim_Amount']]
negative_stay = df[df['Length_of_Stay'] < 0]
negative_days = df[df['Days_Between_Service_and_Claim'] < 0]

print("Invalid age rows:", len(invalid_age))
print("Invalid claim amount rows:", len(invalid_amount))
print("Approved > Claim rows:", len(approved_exceeds))
print("Negative stay rows:", len(negative_stay))
print("Negative days rows:", len(negative_days))
```

```python
df['Prior_Visits_12m'] = df['Prior_Visits_12m'].fillna(0)
df['Provider_Specialty'] = df['Provider_Specialty'].fillna("Unknown")
df['Insurance_Type'] = df['Insurance_Type'].fillna("Unknown")
```

```python
df.describe()
df['Claim_Status'].value_counts()
df['Is_Fraud'].value_counts(normalize=True)  # class imbalance check
```

```python
df.to_csv('../data/healthcare_fraud_detection_cleaned.csv', index=False)
```
